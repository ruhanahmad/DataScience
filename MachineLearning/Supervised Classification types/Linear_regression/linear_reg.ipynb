```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# import online dataset
from sklearn.datasets import load_digits

digits = load_digits()

# input variabe ya feature
digits.data.shape
```

```python
plt.figure(figsize=(20,4))
for index ,(image,label) in enumerate(zip(digits.data[0:10], digits.target[0:10])):
    plt.subplot(1,10,index +1)
    plt.imshow(np.reshape(image,(8,8)), cmap = plt.cm.gray)
    plt.title('Traingin : %i\n' % label,fontsize =10)
```

```python
# split the data and train the model
X = digits.data
y= digits.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0,)
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# now trai the model
from sklearn.linear_model import LogisticRegression
logregmodel = LogisticRegression().fit(X_train,y_train)
logregmodel
```

```python
preditions= logregmodel.predict(X_test[0:10])
preditions= logregmodel.predict(X_test)
preditions
```

```python
# check the score

score = logregmodel.score(X_test,y_test)
score
```

### Confusion Matrix

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_test,preditions)
cm
```

```python
# darw heatmap

import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt = ".3f",linewidths=.5,square=True,cmap='Spectral');
plt.ylabel("Actual Input");
plt.xlabel("predicted output");
all_sample_title = 'Accuracy Score:{0}' .format(score)
plt.title(all_sample_title,size=15)
```
