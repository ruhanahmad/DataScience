```python
def secondApiCall(video_id):

     url_video_stats = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
     response_video_stats = requests.get(url_video_stats).json()
     view_count = response_video_stats['items'][0]['statistics']['viewCount']
     like_count = response_video_stats['items'][0]['statistics']['likeCount']
     comment_count = response_video_stats['items'][0]['statistics']['commentCount']
     favorite_count = response_video_stats['items'][0]['statistics']['likeCount']

     return view_count,like_count,comment_count,favorite_count
```

```python
def apiFirstCall(df):

    for video in apicall["items"]:
        if video['id']['kind'] == "youtube#video":

            video_id = video['id']['videoId']
            video_title = video['snippet']['title']
            video_title = str(video_title).replace("&amp;","")
            video_date = video['snippet']['publishedAt']
            # to split the data
            video_date = str(video_date).split("T")[0]
            video_date      
            view_count,like_count,comment_count,favorite_count =secondApiCall(video_id)
            # save data in pandas df
            df = df.append({
                'video_id':video_id,'video_title':video_title,'video_date':video_date,'view_count':view_count,'like_count':like_count,
                'comment_count':comment_count,'favorite_count':favorite_count
            },ignore_index = True)
    return df
            
   
```

```python

df =pd.DataFrame(columns=["video_id","video_title","video_date","view_count","like_count","comment_count","favorite_count"])
df = apiFirstCall(df)

```
